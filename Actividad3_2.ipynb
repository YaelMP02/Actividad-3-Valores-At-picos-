#Primer hoja

```python
#Carga de un archivo .xlsx desde un libro con varias páginas
DF_2020=pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2020', skiprows=5)
DF_2020
```

```python
#Verificamos informarción del Dataframe
DF_2020.info()
```

```python
valores_nulos=DF_2020.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con 'bfill'
data2=DF_2020.fillna(method='bfill')
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
cualitativas = data2.iloc[ : , [0, 1, 2, 3, 4, 5, 6, 7, 15, 16, 17]]
cuantitativas = data2.iloc[ : , [8, 9, 10, 11, 12, 13, 14, 18]]
```

```python
#Realizamos diagrama de caja o bigote de cada columna dek dataframe
fig = plt.figure(figsize = (15, 8))
cuantitativas.plot(kind='box', vert=False)
plt.title('Valores atipicos del dataframe')
plt.show() #Dibujamos el histograma
```

**PROCEDIMIENTO 'DESVIACIÓN ESTANDAR' PARA ELIMINAR OUTLIERS DE DAATAFRAME**

```python
#Metodo aplicando desviación estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior_iqr= y.mean() + 3*y.std()
Limite_Inferior_iqr= y.mean() - 3*y.std()
print('Limite superior permitido', Limite_Superior_iqr)
print('Limite inferior permitido', Limite_Inferior_iqr)
```

```python
#Encontramos outliers del DataFrame
outliers=cuantitativas[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers
```

**Aqui se convierten los outliers en nulos**

```python
#Obtenemos datos y los outliers se convierten en nulos en el dataframe
data3=cuantitativas[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos (nulos) del dataframe con 'mean'
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(), 1))
data_clean
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Unimos la columna 'DF_2020' con el dataframe
Datos_limpios=pd.concat([data_clean, cualitativas], axis=1)
Datos_limpios
```

```python
#Convertir DataFrame a CSV
Datos_limpios.to_csv("Datos_limpios_2020.csv")
```
