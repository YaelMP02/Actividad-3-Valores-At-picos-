```python
data=pd.read_csv('Data_Sin_Nulos.csv')
data
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data1=data.copy()
```

primero


```python
#Realkizamos diagrama de caja o bigote
#Para la primer columna
fig = plt.figure(figsize = (5, 3))
plt.boxplot(data1['ventas_precios_corrientes'])
plt.title('Outliers de ventas_precios_corrientes')
plt.show() #Dibujamos el histograma
```

```python
#Metodo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data1['ventas_precios_corrientes']

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr=percentile75 - percentile25

Limite_Superior_iqr=percentile75 + 1.5*iqr
Limite_Inferior_iqr=percentile25 - 1.5*iqr
print('Limite superior permitido', Limite_Superior_iqr)
print('Limite inferior permitido', Limite_Inferior_iqr)
```

```python
#Encontramos los outliers
outliers_iqr=data1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Encontramos los outliers
data_clean_VP_corriente=data1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_VP_corriente
```

```python
#Realkizamos diagrama de caja o bigote
fig = plt.figure(figsize = (5, 3))
plt.boxplot(data_clean_VP_corriente['ventas_precios_corrientes'])
plt.title('Outliers de ventas_precios_corrientes')
plt.show() #Dibujamos el histograma
```
